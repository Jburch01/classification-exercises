```python
cat_cols, num_cols = [], []
for col in iris_train.columns:
    if iris_train[col].dtype == 'O':
        cat_cols.append(col)
    else:
        if iris_train[col].nunique() < 10:
            cat_cols.append(col)
        else:
            num_cols.append(col)
```

### 2. Univariate Stats
- For each measurement type (quantitative variable): create a histogram, boxplot, & compute descriptive statistics (using .describe()).
- For each species (categorical variable): create a frequency table and a bar plot of those frequencies.
- Document takeaways & any actions.

```python
explore_cols = cat_cols + num_cols
explore_cols
```

```python
for col in explore_cols:
    if col in cat_cols:
        print(f'Univariate assessment of feature {col}:')
        sns.countplot(data=iris_train, x=col)
        plt.show()
        print(
            pd.concat([iris_train[col].value_counts(),
            iris_train[col].value_counts(normalize=True)],
                 axis=1))
    else:
        print(f'Univariate Feature analysis of feature {col}:')
        plt.hist(iris_train[col])
        plt.show()
        sns.boxplot(data=iris_train, x=col)
        plt.show()
        print(iris_train[col].describe())
    print('------------------------------')
    print('------------------------------')
```
